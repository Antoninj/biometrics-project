```python
from PIL import Image, ImageDraw

from skimage.io import imread,imshow
from skimage.exposure import equalize_adapthist
from skimage.viewer import ImageViewer
from skimage.filters import try_all_threshold,gabor,threshold_otsu,threshold_mean,threshold_minimum,threshold_local
from skimage.morphology import skeletonize, skeletonize_3d, thin
from skimage.util import invert
from skimage.feature import canny

import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
img = imread("data/PNG/1_1.png")
```

```python
imshow(img)
```

```python
eq_img = equalize_adapthist(img)
imshow(eq_img)
```

```python
fig, ax = try_all_threshold(eq_img, figsize=(10, 8), verbose=False)
plt.show()
```

```python
thresh_local = threshold_local(eq_img, 125, 'gaussian')
binary_local = eq_img > thresh_local
```

```python
imshow(binary_local)
```

```python
thinned = invert(thin(binary_local,max_iter= 20))
```

```python
imshow(thinned)
```

```python
skeletonized = invert(skeletonize_3d(binary_local))
```

```python
imshow(skeletonized)
```

```python
edges = canny(img, sigma=2.5)
imshow(invert(edges))
```

```python
filt_real, filt_imag = gabor(img, frequency=0.1)
```

```python
imshow(invert(filt_real))
```
